```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed', lw=2) #label is for legend
plt.plot(x, y_2, label='Y2_values', color = 'blue', lw= 2, linestyle = 'solid')
plt.xlabel('x-values') #
plt.ylabel('y-values')
plt.title('X vs Y') 
plt.grid(alpha=0.25) #darkness of the lines of the grid ranges 0 to 1
plt.legend(loc='best')

plt.show() #optional bc python is interactive but needed if envi is not interactive 

#help(plt) - stolen method of finding out what random shit does 

```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
            xerr= x_err, yerr= y_err,
            color='red', linestyle='dashed',
            marker='o', markerfacecolor='yellow', #marker o is the shape of the point
            ecolor='black',capsize=5,
            label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values', rotation = 0) #makes the y axis horizontal
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc= 'lower right')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x_values = np.linspace(0,50,1000) #have to use the number of points thats already set for x so this has to be at the start
y_values = []
#x_values = np.linspace(0,50,100,retstep = True) stands for returnstep=true tells me the interval 

for i in x_values:
    y_values.append((np.exp(1)**(-i/10))*(np.sin(i)))
#print (x_values)
#print(y_values)

plt.plot(x_values, y_values)
plt.grid(alpha =.25)
plt.xlabel('x', color = 'darkorange',fontsize = 15)
plt.ylabel('y', color = 'darkorange', fontsize = 15, rotation = 0) #i couldnt stand the sideways facing y
plt.title('y=$e^{-x/1/0}sin(x)$', fontsize=20)
plt.ylim(-1,1)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2) #start, stop, step
print(t)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
