```python
import numpy as np 
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]


# for loops are for variable in iterable(eg.list)



co2array = np.array(co2_data)
#print(co2array)
co2array_data = co2array[:,1] #for all elements of co2array picks out the index1 so the conc, not the year
#print(co2array_data)
#co2conclist = list(co2array_data)
#print(type(co2conclist))
#print(co2conclist) #numpy float 

wanted_data = [] #creating a list
increase_data = [] # creating a list 
year = [] #creating a list 
insufficient_increase = []

for i in range(19): #looks at each index individually as opposed to for i in co2data_array, which looks at the elements individually 
    if (co2array_data[i+1] - co2array_data[i]) >= 3 : #i here ends up as an element and not the index 
        wanted_data.append(co2array_data[i+1]) 
        increase_data.append(co2array_data[i+1] - co2array_data[i]) #the amount of increase
        #location.append(i)
        year.append(co2array[i,0])
    else:
        insufficient_increase.append(i)
print(insufficient_increase)
print(increase_data)
print(year)

nice_year = [float(i) for i in year] #list comprehension because flaot only accepts one argument and not a list 
nice_increase_data = [float(j) for j in increase_data]
nice_wanted_data = [float(k) for k in wanted_data]

#the below doesnt work idk why 
#map(float(), year)
#map(float(), wanted_data)
#map(float(), increase_data)

print('in',nice_year,'co2 conc was',nice_wanted_data, 'which is a', nice_increase_data, 'ppm increase from the previous year respectively')
```
